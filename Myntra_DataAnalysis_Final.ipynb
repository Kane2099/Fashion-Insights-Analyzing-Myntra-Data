### Analyzing Address column

```python
import numpy as np

pincode = []
state = []

for i in df['seller_address']:
    
    if i != 'Not Mentioned':
        pincode.append(i.split(' ')[-1])
        state.append(i.split(',')[-1].split('-')[0].strip())
    
    else:
            #Why we are appending for Null values because, then only we will be able to add these values
            # to the df, else it will give number of records not equal error
            pincode.append(np.nan)
            state.append(np.nan)
        
df['pincode'] = pincode
df['state'] = state
```
